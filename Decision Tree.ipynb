Problem Statement:
A cloth manufacturing company is interested to know about the segment or attributes causes high sale.

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import datasets  
from sklearn.model_selection import train_test_split
from sklearn.tree import  DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn import preprocessing
from sklearn import metrics 
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.tree import DecisionTreeRegressor

import warnings
warnings.filterwarnings('ignore')
```

Reading the input csv file

```python
company=pd.read_csv("/content/Company_Data.csv")
company.head()
```

```python
company.describe()                  ## checking the statistics of the data
```

```python
company.info()                 ## check for null values
```

```python
company[company.duplicated(keep = False)]             ##check for duplicate values
```

Checking the presence of outliers using Boxplot

```python
fig, ax = plt.subplots(3, 3, figsize=(15,10))
sns.boxplot(company.Sales, ax=ax[0,0],color='green')
sns.boxplot(company.CompPrice, ax=ax[0,1],color='green')
sns.boxplot(company.Income, ax=ax[0,2],color='green')
sns.boxplot(company.Advertising, ax=ax[1,0],color='green')
sns.boxplot(company.Population, ax=ax[1,1],color='green')
sns.boxplot(company.Price, ax=ax[1,2],color='green')
sns.boxplot(company.Age, ax=ax[2,1],color='green')
sns.boxplot(company.Education, ax=ax[2,2],color='green')

plt.tight_layout()
plt.show()
```

Finding the average sales and classifying the categories that are greater than the average sales value, as the fast moving product segment

```python
company["sales"]=(company["Sales"]>7.49)*1        ##from the stats above avg of sales is 7.49 ().describe)
```

Converting all the categorical variable to numerical values using Label encoding

```python
label_encoder = preprocessing.LabelEncoder()
company['ShelveLoc']= label_encoder.fit_transform(company['ShelveLoc']) 
company['Urban']= label_encoder.fit_transform(company['Urban']) 
company['US']= label_encoder.fit_transform(company['US']) 
```

```python
company
```

Target column is sales(Y), so will drop it

```python
company.drop(["Sales"],axis=1,inplace=True)
```

Seperating the data into independent and target variables

```python
x=company.iloc[:,0:10]
y=company.iloc[:,10]
```

Using Heatmap to check correlation

```python
plt.figure(figsize=(10,7))
sns.heatmap(company.corr()>0.6, annot = True)
plt.show()
```

Correlation matrix

```python
plt.figure(figsize=(10,7))
sns.heatmap(company.corr(),cmap="YlGnBu",annot=True)
plt.show()
```

Splitting data into training and testing data set

```python
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=20)
```

Building Decision Tree classifier model using depth of 3 

```python
model=DecisionTreeClassifier(criterion = 'entropy',max_depth=3)
model.fit(x_train,y_train)
```

Plotting the decision tree

```python
plt.figure(figsize=(10,7))
tree.plot_tree(model);
```

Visualizing how the data is choosing the root node and how a decision tree classifier is built?

```python
fn=['CompPrice','Income','Advertising','Population','Price','ShelveLoc','Age','Education','Urban','US','income']
cn=['YES','NO']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (5,5), dpi=300)
tree.plot_tree(model,
               feature_names = fn, 
               class_names=cn,
               filled = True); 
```

Checking the score for the model (training data)

```python
model.score(x_train, y_train)
```

Predicting on the test data using the built Decision tree classifier model 

```python
pred=model.predict(x_test)
```

Identifying the accuracy of the predicted and the y_test values

```python
np.mean(pred==y_test)
```

Finding the score of the model (test data)

```python
model.score(x_test, y_test)
```

```python
model.feature_importances_
```

```python
y_pred = pred
('Accuracy Score:',metrics.accuracy_score(y_test,y_pred))
```

Predicting the probabilities of each predicted classes

```python
model.predict_proba(x_test)
```

**Confusion matrix** as evaluation metrics

```python
cm= confusion_matrix(y_test, y_pred) 
cm
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(7,5))
sns.heatmap(data=cm,linewidths=.5, annot=True,square = True)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
all_sample_title = 'Accuracy Score: {0}'.format(model.score(x_test, y_test))
plt.title(all_sample_title, size = 15)
```

Printing the Classification report using the Y_test data and the predicted data

```python
print("Classification report - \n", classification_report(y_test,y_pred))
```

Creating a for loop to identify the train and test accuracies of the prediction, with the change in the max_depth feature of the decision tree classifier from 1 to 11.

```python
train_accuracy = []
test_accuracy = []
for max_depth in range(1,11):
    model = DecisionTreeClassifier(criterion='entropy', max_depth= max_depth, random_state= 40)
    model.fit(x_train, y_train)
    train_accuracy.append(model.score(x_train, y_train))
    test_accuracy.append(model.score(x_test, y_test))
```

Listing down the accuracies in a dataframe, for different max depth 

```python
accuracies = pd.DataFrame({'max_depth':range(1,11),'train_acc':train_accuracy, 'test_acc':test_accuracy})
accuracies
```

Visualizing the accuracy graph, based on the max_depth

```python
labels =['train_acc','test_acc']
fig = plt.figure(figsize=(15,6))
ax = fig.add_subplot()
plt.plot(accuracies['max_depth'],accuracies['train_acc'], marker = '*')
plt.plot(accuracies['max_depth'], accuracies['test_acc'], marker = 'o')
ax.set_xticks(np.arange(1,11))
plt.legend(labels)
plt.xlabel('max_depth of the Decision tree')
plt.ylabel('Performance')
plt.show()
```
